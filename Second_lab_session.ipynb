```python
import seaborn as sns
```

```python
sns.regplot(df["BA"], df["W"], ci = None) #ci == Confidence interval 
plt.ylabel('Number of Wins', fontsize = 14) #by setting ci as none, a straight line is drawn
plt.xlabel('Batting Average', fontsize = 14)
plt.title("Number of Wins vs Battling Average", fontsize = 20)
```

```python
plt.scatter(df['RS'], df['W'], s=60) #"s" for  size of data points
sns.regplot(df["RS"], df["W"], ci = None) #ci == Confidence interval
plt.ylabel('Number of Wins', fontsize = 14)
plt.xlabel('Runs Scored', fontsize = 14)
plt.title("Number of Wins vs Runs Scored", fontsize = 20)
```

```python
plt.scatter(df['RA'], df['W'], s=60) #"s" for  size of data points
sns.regplot(df["RA"], df["W"], ci = None) #ci == Confidence interval
plt.ylabel('Number of Wins', fontsize = 14)
plt.xlabel('Runs Allowed', fontsize = 14)
plt.title("Number of Wins vs Runs Allowed", fontsize = 20)
```

```python
#running both the spearman and pearson's correlation
from scipy.stats.stats import pearsonr #Pearson's
from scipy.stats.stats import spearmanr #Spearman's
```

```python
r, p = pearsonr (df['BA'], df['W'])
```

```python
print("r =", round (r,3))
if p < .001:
  print("p < .001")
elif p < .01:
  print("p < .01")
elif p < .05:
  print("p < .05")
else:
  print("Not significant")
```

```python
r, p = pearsonr (df['RS'], df['W'])
print("r =", round (r,3))
if p < .001:
  print("p < .001")
elif p < .01:
  print("p < .01")
elif p < .05:
  print("p < .05")
else:
  print("Not significant")
```

```python
r, p = pearsonr (df['RA'], df['W'])
print("r =", round (r,3))
if p < .001:
  print("p < .001")
elif p < .01:
  print("p < .01")
elif p < .05:
  print("p < .05")
else:
  print("Not significant")
```

```python
#Based on the output, we can say the following
# - there is a statistically significant positive correlation (r = .409, p<.001) between the number of wins and batting average
#- there is a statistically significant positive correlation (r = .512, p<.001) between the number of wins and runs scored
#- there is a statistically significant negative correlation (r= -0.532, p<0.001) between the number of wins and runs allowed

#All these three outputs would reject null hypotheses and support alternative hypotheses based on both p-values and correlation coefficients (effect size)
```
